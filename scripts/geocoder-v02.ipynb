```python
import pandas as pd
from geopandas. tools import geocode
```

```python
df = pd.read_csv('../data/data_preprocessed_v02_geocoding')
```

```python
origins = df['origin_city']
destinations = df['destination_city']
```

```python
origins_unique = origins.unique()
destinations_unique = destinations.unique()
```

```python
origins_df = pd.DataFrame(origins_unique)
destinations_df = pd.DataFrame(destinations_unique)
```

```python
origins_df
```

```python
destinations_df
```

```python
for index,row in origins_df.iterrows():
    try:
        print(row[0])
        info = geocode(row[0],provider='nominatim',user_agent='rnd-geocode-2')
        origins_df.loc[int(index), 'Longitude'] = info['geometry'].loc[0].x
        origins_df.loc[int(index), 'Latitude'] = info['geometry'].loc[0].y
    except Exception as e:
        print(e)
```

```python
origins_df
```

```python
for index,row in destinations_df.iterrows():
    try:
        print(row[0])
        info = geocode(row[0],provider='nominatim',user_agent='rnd-geocode-2')
        destinations_df.loc[int(index), 'Longitude'] = info['geometry'].loc[0].x
        destinations_df.loc[int(index), 'Latitude'] = info['geometry'].loc[0].y
    except Exception as e:
        print(e)
```

```python
destinations_df
```

```python
destinations_df.to_csv('destinations_geocoded.csv')
```

```python
origins_df.to_csv('origins_geocoded.csv')
```
